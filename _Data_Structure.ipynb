# Linked List

```python
linked_list = ['first', 'second', 'third']
print(linked_list)
 
print() 
 
# Menambahkan Element
linked_list.append('fourth') #akan selalu ditambah ke paling belakang
linked_list.append('fifth') #append() takes exactly one argumen
linked_list.insert(3, 'sixth') #akan ditambahkan ke indeks ke 3
linked_list.insert(0, 'zero')
print(linked_list)
print()
 
linked_list.remove('second')
print(linked_list)
print()
```

# Graph

```python
class adjnode:
    def __init__(self, val):
        self.val = val
        self.next = None
 
 
class graph:
    def __init__(self, vertices):
        self.v = vertices
        self.ele = [None]*self.v
 
    def edge(self, src, dest): #src (source), dest (destination)
        node = adjnode(dest)
        node.next = self.ele[src]
        self.ele[src] = node
 
        node = adjnode(src)
        node.next = self.ele[dest]
        self.ele[dest] = node
 
    def __repr__(self):
        for i in range(self.v):
            print("Adjacency list of vertex {}\n head".format(i), end="")
            temp = self.ele[i]
            while temp:
                print(" -> {}".format(temp.val), end="")
                temp = temp.next

g = graph(4)
g.edge(0, 2)
g.edge(1, 3)
g.edge(3, 2)
g.edge(0, 3)
g.__repr__()
 
```

# Hashmaps

```python
def printdict(d):
    for key in d:
        print(key, "->", d[key])
 
 
hashmaps = {0: 'first', 1: 'second', 2: 'third'}
printdict(hashmaps)
print()
 
hashmaps[3] = 'fourth' #menambah elemen baru 
printdict(hashmaps)
print()

hashmaps[4] = 'fifth'
printdict(hashmaps)
print()
 
hashmaps.popitem() #menghilangkan indeks dimulai dr indeks terakhir
printdict(hashmaps)
```
